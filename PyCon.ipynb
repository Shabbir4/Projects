## 9. Find the bad data in the stop_duration column and fix it

```python
ri.stop_duration.value_counts(dropna=False)
```

```python
ri['stop_duration']= ri.stop_duration.fillna('0 mins')
```

```python
ri['stop_duration']= ri.stop_duration.str.replace('0 mins', 'Not stopped')
```

```python
ri.stop_duration.value_counts(dropna=False)
```

```python
ri.stop_duration.str.contains('2')
```

```python
ri.loc[(ri.stop_duration=='1')| (ri.stop_duration=='2'), 'stop_duration']=np.nan
## For applying two conditions, square brackets should be used
## loc to be used for filtering out rows- the first part is for applying condition on rows 
## the second part is for selecting the column
```

```python
ri.stop_duration.value_counts()
```

```python
ri.loc[ri.stop_duration=='Not stopped', 'stop_duration']=np.nan
# np.nan is to be used for filling NaNs
# Data should be checked before removing NaNs, not necessary to remove them every time.
```

```python
ri.stop_duration.value_counts(dropna=False)
```
